```python
# -*- coding: utf-8 -*-
# Initial setup

%precision 3
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Prepare data
t = np.arange(0, 0.99, 0.01)
y1 = np.sin(2 * np.pi * 4 * t)
y2 = np.cos(2 * np.pi * 4 * t)

# Draw scatter plot
plt.scatter(x=t, y=y1)
plt.scatter(x=t, y=y2, c='r')
plt.xlabel('time')
plt.ylabel('value')
plt.legend(['sin', 'cos'], loc='upper left')
plt.show()
```
